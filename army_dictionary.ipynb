```python
##Creating a list of stop words and adding custom stopwords
import codecs
with codecs.open('army_slang.txt', 'r', encoding='utf8') as f:
    text = f.read()
```

```python
preposition_letters = 'ΧΧ•Χ©Χ›ΧΧΧΧ”Χ‘'
```

```python
phrases = text.replace('\r','').split('\n')
```

```python
set_phrases = set()
```

```python
for phrase in phrases:
    set_phrases.add(phrase)
```

```python
f = open('army_slang_.txt','w')
for slang in set_phrases:
    f.write(f'{slang}\n')
f.close()
```
