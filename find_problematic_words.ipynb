```python
##Creating a list of stop words and adding custom stopwords
import codecs
with codecs.open('bgulex.utf8.hr', 'r', encoding='utf8') as f:
    text = f.read()
```

```python
file_type = 'thread'
field_name = 'title'
```

```python
lines = text.split('\n')
```

```python
words_in_heb_dict = set()
for line in lines:
    word = line.split(':')[0].split(' ')[0].strip().replace(' DEF','')
    if len(word) > 0:
        words_in_heb_dict.add(word)
```

```python
##Creating a list of stop words and adding custom stopwords
import codecs
with codecs.open('army_dictionary.txt', 'r', encoding='utf8') as f:
    text = f.read()
```

```python
lines = text.split('\n')
```

```python
words_in_army_dict = set()
for line in lines:
    words_in_army_dict.add(line.replace('\"',""))
```

```python
import pickle
import re
```

```python
pickle_file = open(f'{file_type}_{field_name}_token_lists.pickle', "rb" )
corpus = pickle.load( pickle_file )
pickle_file.close()
```

```python
problem_words = list()
good_words = list()
joined_words = list()
prev_raw_word = ''
for sentence in corpus:
    for raw_word in sentence:
        word = re.sub(r'[!|?|\.|,|-|\(|\)|:|\*|;|\-|\+\/|\'|״]','',raw_word)
        if not word in words_in_heb_dict and not word.isdigit():
            joined_word = str(prev_raw_word+word)
            if len(word) > 1:
                if joined_word in words_in_heb_dict:
                    joined_words.append(joined_word)
                else:
                    if len(prev_raw_word) == 1:
                        problem_words.append(joined_word)
                    else:
                        problem_words.append(word)
        else:
            good_words.append(word)
        prev_raw_word = word
```

```python
print('ratio problematic words')
print(float(len(problem_words))/((len(good_words))+len(joined_words)+len(problem_words)))
```

```python
print('unique ratio problematic words')
print(float(len(set(problem_words)))/((len(set(good_words)))+len(set(joined_words))+len(set(problem_words))))
```

```python
count_army_words = 0
for i in set(problem_words):
    if not i in words_in_army_dict:
        print(i)
```

```python
f = open('problematic_words.txt',"w")
for word in problem_words:
    f.write(f'{word}\n')
f.close()
```
